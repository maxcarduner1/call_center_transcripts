```python
%sql
CREATE OR REPLACE FUNCTION get_call_center_scores_by_rep(rep_id STRING)
RETURNS TABLE
RETURN
  SELECT 
        call_outcome,
        call_purpose,
        scorecard_json,
        total_score,
        transcript_summary
  FROM cmegdemos_catalog.telco_call_center_analytics.call_center_scores
  WHERE rep_id = rep_id;
```

```python
%sql
SELECT * FROM get_call_center_scores_by_rep('REP00000002')
```

```python
%sql
CREATE OR REPLACE TABLE call_center_rep_lookup AS
SELECT DISTINCT
  rep_id,
  rep_name
FROM cmegdemos_catalog.telco_call_center_analytics.call_center_scores
WHERE transcript IS NOT NULL
ORDER BY rep_id
```

```python
%sql
SELECT * FROM call_center_rep_lookup;
```

```python
%sql
CREATE OR REPLACE FUNCTION get_rep_id_by_name(rep_name STRING)
RETURNS STRING
RETURN
  SELECT rep_id
  FROM cmegdemos_catalog.telco_call_center_analytics.call_center_rep_lookup
  WHERE LOWER(rep_name) = LOWER(rep_name)
  LIMIT 1;
```

```python
%sql
SELECT get_rep_id_by_name('Emily Chen')
```
