```python
%sql
CREATE OR REPLACE FUNCTION telco_call_center.analytics.get_call_center_scores_by_rep(rep_id STRING)
RETURNS TABLE
RETURN
  SELECT *
  FROM telco_call_center.analytics.call_center_scores
  WHERE call_center_rep_id = rep_id;
```

```python
%sql
SELECT * FROM telco_call_center.analytics.get_call_center_scores_by_rep('REP00000002')
```

```python
%sql
CREATE OR REPLACE TABLE telco_call_center.analytics.call_center_rep_lookup AS
SELECT DISTINCT
  call_center_rep_id,
  ai_query(
    'databricks-meta-llama-3-3-70b-instruct',
    'Extract only the first name of the call center representative from this transcript. Return only the name, nothing else: ' || transcript
  ) AS rep_name
FROM telco_call_center.analytics.call_center_scores
WHERE transcript IS NOT NULL
ORDER BY call_center_rep_id
```

```python
%sql
SELECT * FROM telco_call_center.analytics.call_center_rep_lookup;
```

```python
%sql
CREATE OR REPLACE FUNCTION telco_call_center.analytics.get_rep_id_by_name(rep_name STRING)
RETURNS STRING
RETURN
  SELECT call_center_rep_id
  FROM telco_call_center.analytics.call_center_rep_lookup
  WHERE LOWER(rep_name) = LOWER(rep_name)
  LIMIT 1;
```
